```sql
%python
dbutils.widgets.removeAll()
```

```sql
create widget text storageName default "datalakejm";
```

```sql
DROP CATALOG IF EXISTS catalog_ut_smartdata CASCADE;
```

```sql
CREATE CATALOG IF NOT EXISTS catalog_ut_smartdata;
```

```sql
CREATE SCHEMA IF NOT EXISTS catalog_ut_smartdata.raw;
CREATE SCHEMA IF NOT EXISTS catalog_ut_smartdata.bronze;
CREATE SCHEMA IF NOT EXISTS catalog_ut_smartdata.silver;
CREATE SCHEMA IF NOT EXISTS catalog_ut_smartdata.golden;
CREATE SCHEMA IF NOT EXISTS catalog_ut_smartdata.exploratory;

CREATE VOLUME IF NOT EXISTS catalog_ut_smartdata.raw.datasets;
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-raw`
URL 'abfss://raw@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación externa para las tablas raw del Data Lake';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-unit-catalog`
URL 'abfss://unit-catalog@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación externa para las tablas unit-catalog del Data Lake';
```
